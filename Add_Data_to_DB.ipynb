```python
#import thư viện kết nối với mysql qua python
import mysql.connector
```

```python
#Lấy dữ liệu từ file csv về, lấy các comment liên quan đến Công Vinh, Thủy Tiên, từ thiện
import csv
data = []
with open ('./python-ws/200k_comments.csv',mode='r', encoding= 'utf8') as file:
    for i in file:
        if "công vinh" in i.lower() or "thuỷ tiên" in i.lower() or "từ thiện" in i.lower():
            data.append(i.strip())
```

```python
# Đẩy dữ liệu vào database 
try:
    conn = mysql.connector.connect(
                host = '127.0.0.1',
                user = 'thinh',
                password = '      ',
                database = 'company'
            )
    cur = conn.cursor()
    cur.execute('Drop table if exists comment1')
    create_script = '''
    CREATE TABLE IF NOT EXISTS comment1 (column1 text(5000)) 
    '''
    #chọn định dạng text vì có nhiều comment rất dài 
    cur.execute(create_script)
    insert_script = 'insert into comment1 values (%s)'
    for i in data:
        cur.execute(insert_script,(i,))
    conn.commit()
except Exception as error:
    print(error)
finally:
    if cur is not None:
        cur.close()
    if conn is not None:
        conn.close()
```

```python
#show dữ liệu trong database
try:
    conn = mysql.connector.connect(
                host = '127.0.0.1',
                user = 'thinh',
                password = '      ',
                database = 'company'
            )  
    cur = conn.cursor()
    cur.execute('select * from comment1')
    for j in cur.fetchall():
        print(j)
    conn.commit()
except Exception as error:
    print(error)
finally:
    if cur is not None:
        cur.close()
    if conn is not None:
        conn.close()
```
