```python
#import thÆ° viá»‡n káº¿t ná»‘i vá»›i mysql qua python
import mysql.connector
```

```python
#Láº¥y dá»¯ liá»‡u tá»« file csv vá», láº¥y cÃ¡c comment liÃªn quan Ä‘áº¿n CÃ´ng Vinh, Thá»§y TiÃªn, tá»« thiá»‡n
import csv
data = []
with open ('./python-ws/200k_comments.csv',mode='r', encoding= 'utf8') as file:
    for i in file:
        if "cÃ´ng vinh" in i.lower() or "thuá»· tiÃªn" in i.lower() or "tá»« thiá»‡n" in i.lower():
            data.append(i.strip())
```

```python
# Äáº©y dá»¯ liá»‡u vÃ o database 
try:
    conn = mysql.connector.connect(
                host = '127.0.0.1',
                user = 'thinh',
                password = '      ',
                database = 'company'
            )
    cur = conn.cursor()
    cur.execute('Drop table if exists comment1')
    create_script = '''
    CREATE TABLE IF NOT EXISTS comment1 (column1 text(5000)) 
    '''
    #chá»n Ä‘á»‹nh dáº¡ng text vÃ¬ cÃ³ nhiá»u comment ráº¥t dÃ i 
    cur.execute(create_script)
    insert_script = 'insert into comment1 values (%s)'
    for i in data:
        cur.execute(insert_script,(i,))
    conn.commit()
except Exception as error:
    print(error)
finally:
    if cur is not None:
        cur.close()
    if conn is not None:
        conn.close()
```
